## Imports:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Initial EDA:

```python
all_data = pd.read_csv('all_data.csv') #GDP and Life Expectancy data from 2000 to 2015 for 6 countries
print(all_data.head())
print()
print(all_data.info()) #all_data.Country is listed as an object, probably some data issues to clean out
print()
print(all_data.Country.value_counts()) #no issues found, just not listed as a string for some reason

#let's rename the life expectancy column while we're here
all_data = all_data.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")

#and let's rename the US to somethinga bit shorter
all_data.replace('United States of America', 'US', inplace=True)
```

```python
#Let's investigate the range of the data a bit

Countries_List = all_data.Country.unique()

#GDP Values Printout
print('GDP Summary (in Trillions)')
GDP_Dictionary = {}
for i in range(len(Countries_List)):
    country_min = all_data[all_data.Country == Countries_List[i]].GDP.min()/1000000000000
    country_max = all_data[all_data.Country == Countries_List[i]].GDP.max()/1000000000000
    country_range = country_max - country_min
    print(f'{Countries_List[i]} has a max of {country_max} and a min of {country_min}: Range = {country_range}')

#Life Expectancy Values Printout
print('\nLife Expectancy Summary')
LEABY_Dictionary = {}
for i in range(len(Countries_List)):
    country_min = all_data[all_data.Country == Countries_List[i]].LEABY.min()
    country_max = all_data[all_data.Country == Countries_List[i]].LEABY.max()
    country_range = country_max - country_min
    print(f'{Countries_List[i]} has a max of {country_max} and a min of {country_min}: Range = {country_range}')
```

## Distribution Analysis of GDP and Life Expectancy

```python
#I'd also like to take a look at a few distribution views of the GDP and Life Expectancy Data to analyze the skewing a bit further

plt.figure(figsize=[12,8])
sns.displot(all_data.GDP, alpha=.5)
plt.title('GDP Distribution')
plt.show()
plt.clf()

plt.figure(figsize=[12,8])
sns.displot(all_data.LEABY, alpha=.5)
plt.title('Life Expectancy Distribution')
plt.show()
plt.clf()

#From this, we see that GDP is noteably right skewed and Life Expectancy is noteably left skewed
```

## Life Expectancy (Part 1)

```python
#Let's look at the boxplots for life expectancy per country
plt.figure(figsize=[10,15])
sns.boxplot(y='LEABY', x='Country', data=all_data)
plt.ylabel('Life expectancy at birth (years)')
#plt.xticks(rotation=30)... turns out I didn't need this once I went back to re-label the US values
plt.show()
plt.clf()
#and again with Zimbabwe in a category all it's own
```

## GDP vs. Life Expectancy (Part 1)

```python
#Let's do 1 figure with a Life Expectancy chart on the left, and then a GDP chart to the right
plt.figure(figsize = [20,10])
plt.subplot(1,2,1)
sns.lineplot(x='Year', y='LEABY', data=all_data, hue='Country')
plt.title('Life Expectancy by Year')

plt.subplot(1,2,2)
sns.lineplot(x='Year', y='GDP', data=all_data, hue='Country')
plt.title('GDP by Year')
plt.show()
plt.clf()
```

```python
#Now we will use scatter plots to check for positive or negative correlation between GDP and Life Expectancy
#We'll do 1 combined scatter plot first to see if there is a general trend
plt.figure(figsize = [10,10])
sns.scatterplot(x='GDP', y='LEABY', data=all_data, hue='Country')
plt.title('GDP vs. Life Expectancy (All Countries)')
plt.show()
plt.clf()
```

The data clearly indicates that we may be dealing with 3 separate categories of GDP size 
- Chile and Zimbabwe GDP is on the low end with very low GDP Growth
- Mexico and Germany are in the middle with intermediate GDP Growth
- China and United States are on the high end with rapid GDP Growth

## GDP vs. Life Expectancy (Part 2)

```python
#Now we will make a single scatter per Country, and also combined scatters per each of the 3 country sets as proposed earlier
#plt.figure(figsize = [10,15])
#This got a bit tricky, as Seaborn does not cooperate with subplots for lmplot function (no ax parameter)
#Therefore, I split this into 3 data sets so I can do 3 separate column separations based on Country... 
#But then the col feature still didn't look that great in comparison so I did 3 charts separated by hue

Chile_Zimbabwe = all_data[(all_data.Country == 'Chile') | (all_data.Country == 'Zimbabwe')]

sns.lmplot(x='GDP', y='LEABY', data=Chile_Zimbabwe, hue='Country', fit_reg=True)
plt.title('GDP vs. Life Expectancy (Chile and Zimbabwe)')
plt.ylabel('Life expectancy at birth (years)')

Mexico_Germany = all_data[(all_data.Country == 'Mexico') | (all_data.Country == 'Germany')]

sns.lmplot(x='GDP', y='LEABY', data=Mexico_Germany, hue='Country', fit_reg=True)
plt.title('GDP vs. Life Expectancy (Mexico and Germany)')
plt.ylabel('Life expectancy at birth (years)')

China_US = all_data[(all_data.Country == 'China') | (all_data.Country == 'US')]

sns.lmplot(x='GDP', y='LEABY', data=China_US, hue='Country', fit_reg=True)
plt.title('GDP vs. Life Expectancy (China and US)')
plt.ylabel('Life expectancy at birth (years)')

plt.show()
plt.clf()
```

Main learning hear is that Zimbabwe seems to be in a category of it's own compared to the rest due to low / compressed GDPs and much lower than average life expectancy

## GDP vs. Life Expectancy (Part 3b)

```python
#Let's take a look at the data spread using Violin Plots instead of Box and Whisker, maybe it will add a bit of nuance
#This is also my first time trying to use a shared axes on two plots, so rest assured I compiled this section at least 10 times while I got it right
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(20, 10), gridspec_kw={'width_ratios': [11,10]})
axes[0] = sns.boxplot(ax = axes[0], y='Country', x='GDP', data=all_data)
axes[0].set_xlabel('GDP per Country')
axes[1] = sns.violinplot(ax = axes[1], y='Country', x='LEABY', data=all_data)
axes[1].set_xlabel('Life Expectancy per Country')
```
